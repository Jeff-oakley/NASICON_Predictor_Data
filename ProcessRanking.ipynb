## Processing data from ranking model

```python
#!/usr/bin/env python

import os
from monty.serialization import loadfn,dumpfn


Prefix=os.getcwd();
MinInd=1; MaxInd=1000000;
NModelPerJob=8000;

NJob=int(MaxInd/NModelPerJob);
StaJobInd=0; i=0;
ModelDict={'svm':{},'tree':{}};

for Model in ['svm','tree']:
    for Ind in range(NJob):
        StaInd=Ind*8000; FinInd=(Ind+1)*8000-1;
        MsonName='Model{}InfoStaInd{}FinInd{}.mson'.format(Model,StaInd,FinInd);
        print('Processing {}'.format(MsonName));
        if not os.path.isfile(MsonName): print('{} is not found'.format(MsonName)); continue;
        Dict=loadfn(MsonName);
        ModelDict[Model].update(Dict);
        i+=1;
    
    StaInd=(Ind+1)*8000; FinInd=MaxInd;
    MsonName='Model{}InfoStaInd{}FinInd{}.mson'.format(Model,StaInd,FinInd);
    print('Processing {}'.format(MsonName));
    if not os.path.isfile(MsonName): print('{} is not found'.format(MsonName)); continue;
    Dict=loadfn(MsonName);
    ModelDict[Model].update(Dict)
    i+=1;
    
dumpfn(ModelDict,'AllModels.mson');
```

```python
KeyTest=list(ModelDict['svm'].keys())[0];
print(ModelDict['svm'][KeyTest]);
ModelsvmLst=[Value for Key,Value in ModelDict['svm'].items()];
ModeltreeLst=[Value for Key,Value in ModelDict['tree'].items()];
```
